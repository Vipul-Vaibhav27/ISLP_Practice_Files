# We can create more sohisticated plots, using ax.contour() method which produces a contour plot in order to represent 3-D data, similar to a topographical map

contour method takes three arguments
1) a vector of x values(first dimension)
2) a vector of y values(second dimension)
3) a matrix whose elements correspond to the z value(the third dimension) for each pair of (x, y) coordinates

to create x and y, we will use the command np.linspace(a, b, n) which returns a vector
of n numbers starting at "a" and ending at "b"

```python
fig, ax = plt.subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1/(1+x**2))
ax.contour(x, y, f);
```

# 2.3.6 Indexing Data

```python
A = np.arange(16).reshape((4, 4))
A
```

```python
A[[1, 3]]  # selects second and fourth row
```

```python
# to select first and third column
A[:, [0, 2]]
```

```python
# Now, there intersection
A[[1, 3], [0, 2]]
```

```python
np.array([A[1, 0], A[3, 2]]) # just their intersections
```

We can see what has gone wrong here. When supplied with two indexing
lists, the numpy interpretation is that these provide pairs of i, j indices for
a series of entries. That is why the pair of lists must have the same length.
However, that was not our intent, since we are looking for a submatrix

```python
A[[1, 3], [0, 2, 3]]
```

One easy way to do this is as follows. We first create a submatrix by
subsetting the rows of A, and then on the fy we make a further submatrix
by subsetting its columns.

```python
A[[1, 3]][:, [0, 2]]
```

```python
# the convenience function np.ix_() allows us to extract a submatrix using lists, by
# creating an intermediate mesh object
idx = np.ix_([1, 3], [0, 2, 3])
A[idx]
```

```python
A[1:4:2, 0:3:2] # 2 and 2 are step sizes
```

##Boolean Indexing

```python
keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
# We now set two of the elements to be true
keep_rows[[1, 3]]=True
keep_rows
```

```python
np.all(keep_rows==np.array([0, 1, 0, 1]))  # np.all() checks if all elements are true
```

```python
# even though they are equal, they retrieves different set of rows
A[np.array([0, 1, 0, 1])] # it retrieves, first and second row twice
```

```python
A[keep_rows] # retrieves only those rows for which the Boolean equals TRUE
```

Above code blocks show that Booleans and integers are treated differently by numpy

```python
# we again creat a mesh containing the second and fourth rows and the first,third,fourth cols
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]]=True
idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]
```

```python
# We can also mix a list with an array of booleans in the arguments to np.ix_()
idx_mixed = np.ix_([1, 3], keep_cols)
A[idx_mixed]
```

### More on selecting rows and columns

```python
# We want a data frame consisting of the "weight" and "origin" of the subset of cars with
# year greater than 80, i.e. those built after 1980
# Therefore, we first create boolean array that indexes the rows
# The loc[] method allows for boolean entries as well as strings
idx_80 = Auto_re['year'] > 80
Auto_re.loc[idx_80, ['weight', 'origin']]
```

```python
# To do this more concisely, we can use an anonymous function called lambda:
Auto_re.loc[lambda df:df['year']>80, ["weight", "origin"]]
# lambda takes df as argument and returns df["year"]>80
```

**Don't use "and" and "or", instead use "&" and "|" respectively since "&" and "|" do element wise operations and "and" and "or" will throw errors**

```python
Auto_re.loc[lambda df: (df['year'] > 80) & (df['mpg'] > 30),
['weight', 'origin']
]
```
