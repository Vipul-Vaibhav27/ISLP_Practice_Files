```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

# 12(a)

For them to be equal,
summation of xi^2 and summation of yi^2 need to be equal,
i.e,
np.dot(x, x)==np.dot(y, y)

# 12(b)

```python
# for them not need to be equal, we have to not satisfy the above condition
# taking rng
rng = np.random.default_rng(69)
x = rng.random(100)
y = 10*x + rng.normal(size=100)
```

```python
x, y
```

```python
result1 = sm.OLS(y, x).fit()
result1.summary()
```

```python
result1_new = sm.OLS(x, y).fit()
result1_new.summary()
```

### Their coefficient estimates are substantially different, i.e 9.7811 and 0.0990.
